```python
# Import necessary libraries
!pip install pillow
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt


```

```python



def encrypt_image(image, key):
    encrypted_image = image.copy()
    pixels = np.array(encrypted_image)
    pixels = pixels ^ key
    encrypted_image = Image.fromarray(pixels)
    return encrypted_image

def decrypt_image(image, key):
    decrypted_image = image.copy()
    pixels = np.array(decrypted_image)
    pixels = pixels ^ key
    decrypted_image = Image.fromarray(pixels)
    return decrypted_image

```

```python

response = requests.get('https://picsum.photos/200/300')
image = Image.open(BytesIO(response.content))

```

```python

plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()

```

```python

key = 123
encrypted_image = encrypt_image(image, key)


```

```python

plt.imshow(encrypted_image)
plt.title("Encrypted Image")
plt.axis('off')
plt.show()


```

```python

decrypted_image = decrypt_image(encrypted_image, key)
```

```python

plt.imshow(decrypted_image)
plt.title("Decrypted Image")
plt.axis('off')
plt.show()
```
